```python
X = df.drop(columns=['left'])  # Bağımsız değişkenler
y = df['left']  # Hedef değişken

smote = SMOTE(sampling_strategy=0.5, random_state=42)  # Azınlık sınıfını artır
X_resampled, y_resampled = smote.fit_resample(X, y)

print("Önceki Sınıf Dağılımı:", Counter(y))
print("SMOTE Sonrası Sınıf Dağılımı:", Counter(y_resampled))
```

```python


# Veriyi %80 eğitim, %20 test olarak ayır
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

print("Eğitim veri seti boyutu:", X_train.shape)
print("Test veri seti boyutu:", X_test.shape)
```

```python
"Random Forest"
# Modeli tanımla
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Modeli eğit
rf_model.fit(X_train, y_train)

# Tahmin yap
y_pred_rf = rf_model.predict(X_test)

# Sonuçları göster
print("Random Forest Sonuçları:")
print(classification_report(y_test, y_pred_rf))
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
```

```python
"XGBoost"
# Modeli tanımla
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)

# Modeli eğit
xgb_model.fit(X_train, y_train)

# Tahmin yap
y_pred_xgb = xgb_model.predict(X_test)

# Sonuçları göster
print("XGBoost Sonuçları:")
print(classification_report(y_test, y_pred_xgb))
print("Accuracy:", accuracy_score(y_test, y_pred_xgb))
```

```python
df = pd.get_dummies(df, columns=['sales'], drop_first=True)
```

```python
# Hiperparametreler için arama alanı
param_dist = {
    'n_estimators': [100, 200, 300, 400],
    'max_depth': [5, 10, 15, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

# Randomized Search
rf_random = RandomizedSearchCV(estimator=RandomForestClassifier(random_state=42),
                               param_distributions=param_dist,
                               n_iter=20, cv=5, verbose=2, n_jobs=-1, scoring='f1')

# Optimizasyon için eğit
rf_random.fit(X_train, y_train)

# En iyi parametreleri yazdır
print("En iyi parametreler:", rf_random.best_params_)
```

```python
# En iyi parametrelerle Random Forest modeli oluştur
final_rf = RandomForestClassifier(
    n_estimators=400,
    min_samples_split=2,
    min_samples_leaf=2,
    max_depth=None,
    bootstrap=True,
    random_state=42
)

# Modeli eğit
final_rf.fit(X_train, y_train)

# Test seti üzerinde tahmin yap
y_pred_final = final_rf.predict(X_test)

# Modelin performansını değerlendir
from sklearn.metrics import classification_report, accuracy_score

print("Final Random Forest Modeli Sonuçları:")
print(classification_report(y_test, y_pred_final))
print("Accuracy:", accuracy_score(y_test, y_pred_final))
```
